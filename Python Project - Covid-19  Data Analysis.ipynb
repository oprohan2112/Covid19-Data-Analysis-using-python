### Q. 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
data.head(5)
```

```python
data.groupby("Region").sum()
```

```python
data.groupby("Region")["Recovered"].sum().head(10).sort_values(ascending = False)
```
